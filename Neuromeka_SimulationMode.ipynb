## 외부 제어 허용

```python
def get_operation_mode_config():
    response = indy.config.GetOperationModeConfig(common_msgs.Empty())
    return MessageToDict(response,
                         including_default_value_fields=True,
                         preserving_proto_field_name=True,
                         use_integers_for_enums=True)

def set_operation_mode_config(config: dict):
    req = config_msgs.OperationModeConfig()
    ParseDict(config, req)
    response = indy.config.SetOperationModeConfig(req)
    return MessageToDict(response,
                         including_default_value_fields=True,
                         preserving_proto_field_name=True,
                         use_integers_for_enums=True)

# 사용
config = get_operation_mode_config()
print(config)
```

```python
# 1. 현재 config 가져오기
config = get_operation_mode_config()

# 2. 원하는 값 변경
config['use_ext_auto_mode'] = True  # 또는 True

# 3. 다시 적용
result = set_operation_mode_config(config)

print(result)
```
