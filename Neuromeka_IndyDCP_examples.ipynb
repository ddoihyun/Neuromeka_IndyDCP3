```python
from neuromeka import IndyDCP3
```

```python
indy = IndyDCP3(robot_ip = '192.168.0.161', index=0) #STEP PC IP 주소 입력
```

```python
robot_data = indy.get_robot_data()

# robot_data 딕셔너리 구조 설명

# running_mins       : 로봇 전원 ON 이후 누적 동작 시간 (분)
# running_secs       : 로봇 전원 ON 이후 누적 동작 시간 (초)
# running_hours      : 로봇 전원 ON 이후 누적 동작 시간 (시간)
#
# op_state           : 로봇 동작 상태 코드
#                      (아래 쉘 참고)
#
# q                  : 현재 각 관절(Joint 1~6)의 각도 값 [deg]
#
# qdot               : 각 관절(Joint 1~6)의 각속도 [deg/s]
#
# p                  : TCP(Tool Center Point)의 현재 위치 및 자세
#                      [x, y, z, rx, ry, rz] mm, deg
#
# pdot               : TCP의 선속도 및 각속도
#                      [vx, vy, vz, wx, wy, wz]
#
# ref_frame          : 기준 좌표계 설정 값
#                      (보통 [0,0,0,0,0,0]이면 Base Frame)
#
# tool_frame         : 툴 좌표계 오프셋 값
#
# tool_link          : 적용된 툴 링크 번호
#                      (-1이면 툴 미설정)
#
# locked_joint       : 현재 잠겨 있는 관절 번호
#                      (-1이면 잠긴 관절 없음)
#
# sim_mode           : 시뮬레이션 모드 여부
#                      (True: 시뮬레이터, False: 실제 로봇)

print(robot_data)
```

```python
control_state = indy.get_control_state()

# control_state 딕셔너리 구조 설명 (제어 상태)
#
# ───────────── Joint Space (관절 공간) ─────────────
# q             : 현재 관절 각도 [deg] (Joint 1~6)
# qdot          : 현재 관절 각속도 [deg/s]
# qddot         : 현재 관절 각가속도 [deg/s^2]
#
# qdes          : 목표(Desired) 관절 각도 [deg]
# qdotdes       : 목표 관절 각속도 [deg/s]
# qddotdes      : 목표 관절 각가속도 [deg/s^2]
#
# ───────────── Cartesian Space (작업 공간) ─────────────
# p             : 현재 TCP 위치 및 자세
#                 [x, y, z, rx, ry, rz]mm, deg
#
# pdot          : TCP 선속도 및 각속도
# pddot         : TCP 선가속도 및 각가속도
#
# pdes          : 목표 TCP 위치 및 자세
# pdotdes       : 목표 TCP 속도
# pddotdes      : 목표 TCP 가속도
#
# ───────────── Torque / Force ─────────────
# tau           : 제어기에서 계산된 관절 토크 명령값 [Nm]
# tau_act       : 실제 액추에이터에 인가된 관절 토크 [Nm]
# tau_ext       : 외력에 의해 추정된 관절 토크 [Nm]
#
# tau_jts       : 관절 토크 센서 기반 토크 값 [Nm]
# tau_jts_raw1  : 관절 토크 센서 Raw 값 (ADC/엔코더 값)
# tau_jts_raw2  : 관절 토크 센서 Raw 값 (이중 센서)

print(control_state)
```

```python
motion_data = indy.get_motion_data()

# motion_data 딕셔너리 구조 설명 (모션 / 궤적 상태)
#
# traj_progress       : 현재 모션의 전체 궤적 진행률 [%] (0 ~ 100)
#
# cur_traj_progress   : 현재 실행 중인 궤적 구간의 진행률 [%]
#
# traj_state          : 궤적 상태 코드
#                       0 : 정상 / 완료
#                       (기타 값은 모션 상태 에러 또는 진행 중 상태)
#
# speed_ratio         : 모션 속도 비율 [%] (100 = 설정된 최대 속도)
#
# remain_distance     : 목표점까지 남은 거리 (단위: mm)
#
# is_in_motion        : 현재 모션 실행 중 여부
#                       True  : 이동 중
#                       False : 정지 상태
#
# is_target_reached   : 목표 위치 도달 여부
#                       True  : 목표점 도달
#                       False : 아직 도달 안 함
#
# is_pausing          : 일시정지(Pause) 상태 여부
#
# is_stopping         : 정지(Stop) 명령 처리 중 여부
#
# has_motion          : 실행할 모션이 존재하는지 여부
#
# motion_id           : 현재 또는 마지막 모션 ID
#
# motion_queue_size   : 모션 큐에 대기 중인 모션 개수

print(motion_data)
```
